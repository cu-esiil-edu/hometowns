```python
import os

import cartopy.crs as ccrs
import geopandas as gpd
import holoviews as hv
import hvplot.pandas
import pandas as pd
```

```python
hometowns_df = pd.read_csv(os.path.join('..', 'hometowns.csv'))
hometowns_df
```

```python
hometowns_gdf = gpd.GeoDataFrame(
    hometowns_df,
    geometry=gpd.points_from_xy(
        hometowns_df.longitude, 
        hometowns_df.latitude,
        crs=4326)
)
hometowns_gdf
```

```python
hometowns_plot = hometowns_gdf.hvplot(
    geo=True, tiles='EsriNatGeo', color='red',
    hover_cols=['type_of_place', 'description'],
    width=600, height=600)
hv.save(hometowns_plot, os.path.join('..', 'maps', 'short-course-2024.html'))
```
